```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

Data = pd.read_csv('data/cox-violent-parsed.csv', skiprows = False)
plot = sns.catplot(x="v_score_text", y="v_decile_score", data = Data)

plt.title("Violent Scores Associated with Violent Decile Scores")
plt.xlabel("Danger Level of a Person's Violent Score")
plt.ylabel("Decile Risk Score Level Assigned to Corresponding Violent Score (from 1 to 10)")
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd


Data = pd.read_csv('data/cox-violent-parsed.csv', skiprows = False)
plot = sns.catplot(x="race", y="age", data = Data)

plt.title(f"Ages of Different Ethnic Prisoners")
plt.xticks(rotation=60)
plt.xlabel("Different Ethnicities of People in Prison")
plt.ylabel("Age")
plt.figure(figsize=(20,50))
```
